```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt

sns.set_style("darkgrid")
```

```python
df = pd.read_csv('data.csv', index_col=0)
df.columns = ['n servers', 'gossip rounds', 'uses seeds']
```

```python
plt.figure()
g = sns.relplot(x="n servers", y="gossip rounds", hue="uses seeds", kind="line", data=df, height=2.5, aspect=1.8)
plt.savefig('gossip_scaling.pdf')
```
